# Convergence plot

This collects the accuracy and loss curves of each model and plots them together

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set_context("paper", font_scale=1.6)
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ['Computer Modern Roman']})
```

```python
SHL18_loss = np.load("../../datasets/SHL18/train_loss_results.npy")
SHL18_acc = np.load("../../datasets/SHL18/train_acc_results.npy")
RW16_loss = np.load("../../datasets/RW16/train_loss_results.npy")
RW16_acc = np.load("../../datasets/RW16/train_acc_results.npy")
UCI13_loss = np.load("../../datasets/UCI13/train_loss_results.npy")
UCI13_acc = np.load("../../datasets/UCI13/train_acc_results.npy")
```

```python

fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(8,10))
axs[0].plot(np.arange(1,len(SHL18_acc)+1), SHL18_acc, '--')
#axs[0].plot(np.arange(1,len(SHL20_acc)+1), SHL20_acc, '--')
axs[0].plot(np.arange(1,len(RW16_acc)+1), RW16_acc, '--')
axs[0].plot(np.arange(1,len(UCI13_acc)+1), UCI13_acc, '--')
axs[0].set_xlabel("epoch")
axs[0].set_ylabel("accuracy")
axs[0].legend(["SHL18", "RW16", "UCI16"])
axs[0].grid(axis="y", alpha=0.3)
axs[0].set_xlim([1,np.max([len(SHL18_acc),len(RW16_acc)])])
axs[0].set_ylim([np.clip(np.min([SHL18_acc.min(), RW16_acc.min(), UCI13_acc.min()]),0,1), 1])
axs[0].set_title("Accuracy", fontsize=20)

axs[1].plot(np.arange(1,len(SHL18_loss)+1), SHL18_loss, '--')
#axs[1].plot(np.arange(1,len(SHL20_loss)+1), SHL20_loss, '--')
axs[1].plot(np.arange(1,len(RW16_loss)+1), RW16_loss, '--')
axs[1].plot(np.arange(1,len(UCI13_loss)+1), UCI13_loss, '--')
axs[1].set_xlabel("epoch")
axs[1].set_ylabel("loss")
axs[1].legend(["SHL18", "RW16", "UCI16"])
axs[1].grid(axis="y", alpha=0.3)
axs[1].set_xlim([1,np.max([len(SHL18_loss),len(RW16_loss)])])
axs[1].set_ylim([0, np.max([SHL18_loss.max(), RW16_loss.max(), UCI13_loss.max()])])
axs[1].set_title("Loss", fontsize=20)
plt.tight_layout()
plt.savefig("../../figures/training_figure.pdf")
plt.show()
```
